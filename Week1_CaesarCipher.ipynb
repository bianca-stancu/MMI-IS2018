```python
# Setup the alphabet
AS = AlphabeticStrings()
```

```python
#Initialize a shift crypto system 
caesar = ShiftCryptosystem(AS)
K = 3
```

```python
#Encoding of plaintext to remove any characters that are not alphabetic
plaintext = AS.encoding("Secret message from backstabbed roman dude.")
print("Plain text: " + str(plaintext))
```

```python
#Encrypt the plaintext
ciphertext = caesar.enciphering(K,plaintext)
print("Cipher text: " + str(ciphertext))
```

```python
#Decryption
decrypted = caesar.deciphering(K,ciphertext)
"Decryption successful" if decrypted == plaintext else "Decryption was not successful"
```
